```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def adjust_temperature(probabilities, temperature):
    adjusted_probabilities = np.power(probabilities, 1 / temperature)
    return adjusted_probabilities / np.sum(adjusted_probabilities)
```

```python
mu = 0  # Moyenne
sigma = 1  # Écart-type

# Générer des données selon la distribution gaussienne
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Normaliser les probabilités
probabilities = y / np.sum(y)
```

```python
# Ajuster les probabilités pour différentes températures
T1 = 1
T2 = 2
T3 = 0.5

probs_T1 = adjust_temperature(probabilities, T1)
probs_T2 = adjust_temperature(probabilities, T2)
probs_T3 = adjust_temperature(probabilities, T3)
```

```python
# Visualisation
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].bar(np.arange(len(probabilities)), probs_T1)
axes[0].set_title(f'T = {T1}')
axes[0].set_ylim([0, max(max(probs_T1), max(probs_T2), max(probs_T3))])

axes[1].bar(np.arange(len(probabilities)), probs_T2)
axes[1].set_title(f'T = {T2}')
axes[1].set_ylim([0, max(max(probs_T1), max(probs_T2), max(probs_T3))])

axes[2].bar(np.arange(len(probabilities)), probs_T3)
axes[2].set_title(f'T = {T3}')
axes[2].set_ylim([0, max(max(probs_T1), max(probs_T2), max(probs_T3))])

plt.show()
```
